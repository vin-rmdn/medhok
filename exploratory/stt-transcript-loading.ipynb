```python
# Get all .eaf files
from pathlib import Path

eaf_files = list(Path('../data/audio').glob('**/*.eaf'))
print(f'There are {len(eaf_files)} file{"s" if len(eaf_files) > 1 else ""}.')
```

```python
def extract_transcripts(eaf_files):
    dfs = []
    for i, eaf_file in enumerate(eaf_files):
        filepath = []
        transcript_data = []
        transcript_id = []
        speaker = []
        print('Processing %s (%d/%d)' % (eaf_file, i+1, len(eaf_files)))
        eaf_data = pympi.Elan.Eaf(eaf_file)
        key_interest = [key for key in eaf_data.tiers if 'tx' in key]
        print(f'\tKey interests: {key_interest}')
        for j, key in enumerate(key_interest):
            print(f'\tProcessing speaker {eaf_data.tiers[key][2]["PARTICIPANT"]} ({j + 1}/{len(key_interest)})', end='\r')
            for annot_key, annot_value in eaf_data.tiers[key][0].items():
                filepath.append(eaf_file)
                transcript_id.append(annot_key)
                transcript_data.append(annot_value)
                speaker.append(eaf_data.tiers[key][2]['PARTICIPANT'])
                
        # Dataframe
        _df = pd.DataFrame()
        _df['transcript_id'] = transcript_id
        _df['filepath'] = filepath
        _df['speaker'] = speaker
        _df = pd.concat([
            _df, pd.DataFrame(transcript_data).drop(3, axis=1).rename(columns={
                0: 'ms_start',
                1: 'ms_fin',
                2: 'transcript'
            })], axis=1
        )
        
        _df['ms_start'] = _df['ms_start'].apply(lambda x: eaf_data.timeslots[x])
        _df['ms_fin'] = _df['ms_fin'].apply(lambda x: eaf_data.timeslots[x])
        
        _df['s_start'] = _df['ms_start'].apply(lambda x: x / 1000)
        _df['s_fin'] = _df['ms_fin'].apply(lambda x: x / 1000)
        dfs.append(_df)
        print()
    
    # Merging dataframes together
    df = pd.concat(dfs).reset_index(drop=True)
    return df

transcript_df = extract_transcripts(eaf_files)
```

```python
n_transcript_raw = len(transcript_df)
print(f'There are {n_transcript_raw} raw transcript{"s" if n_transcript_raw > 1 else ""}.')
transcript_df.head()
```

```python
# Extracting audio files according to the transcripts
import librosa
import soundfile as sf

SAMPLE_RATE=16000

# Extracting segments of audio files according to the dataframe to a new wave file
wav_files = transcript_df['filepath'].unique()
for i, wav_file in enumerate(wav_files):
    print('Processing %s (%d/%d)' % (wav_file, i+1, len(wav_files)))
    wav = librosa.load(wav_file, sr=SAMPLE_RATE)[0]
    for j, transcript in transcript_df.loc[transcript_df['filepath'] == wav_file].reset_index(drop=True).iterrows():
        print(f'\tProcessing segment {j + 1}/{len(transcript_df.loc[transcript_df["filepath"] == wav_file])}', end='\r')
        wav_segment = wav[int(SAMPLE_RATE * transcript['s_start']):int(SAMPLE_RATE * transcript['s_fin'])]
        sf.write(f"../data/speech/{transcript['transcript_id']}.wav", wav_segment, SAMPLE_RATE)
    print()
```
